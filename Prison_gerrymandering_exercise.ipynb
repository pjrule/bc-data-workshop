```python
%config InlineBackend.figure_formats = ['svg']
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt
```

```python
no_realloc_gdf = gpd.read_file("pa_vtd20_no_realloc.zip").set_index("GEOID20")
```

```python
realloc_gdf = gpd.read_file("pa_vtd20_realloc.zip").set_index("GEOID20")
```

```python
voter_cols = {
    "Democratic": "democratic_voters",
    "Republican": "republican_voters",
    "Libertarian": "libertarian_voters",
    "Green": "green_voters",
    "NoAffiliation": "no_affiliation_voters",
    "Other": "other_party_voters",
    "Total": "total_voters",
}
```

```python
voter_reg_senate_df = pd.read_csv("pa_voter_registration_senate.csv").set_index("DistrictCode").rename(columns=voter_cols)
```

```python
voter_reg_senate_df
```

```python
voter_reg_congress_df = pd.read_csv("pa_voter_registration_congress.csv").set_index("DistrictCode").rename(columns=voter_cols)
```

```python
realloc_gdf.head()
```

```python
voter_reg_senate_df.head()
```

```python
realloc_pops_senate_df = realloc_gdf.groupby("senate").sum().join(voter_reg_senate_df)
```

```python
no_realloc_pops_senate_df = no_realloc_gdf.groupby("senate").sum().join(voter_reg_senate_df)
```

```python
realloc_pops_senate_df.head()
```

```python
realloc_pops_congress_df = realloc_gdf.groupby("congress").sum().join(voter_reg_congress_df)
```

```python
no_realloc_pops_congress_df = no_realloc_gdf.groupby("congress").sum().join(voter_reg_congress_df)
```

```python
realloc_pops_congress_df
```

```python
no_realloc_pops_congress_df
```

## Let's make a plot!

```python
pop_diff_senate_df = realloc_pops_senate_df - no_realloc_pops_senate_df
```

```python
fig, ax = plt.subplots()
pop_shift_pct = 100 * pop_diff_senate_df['total_pop'] / no_realloc_pops_senate_df['total_pop']
dem_share_pct = 100 * realloc_pops_senate_df['democratic_voters'] / realloc_pops_senate_df['total_voters']
ax.scatter(pop_shift_pct, dem_share_pct)
ax.set_xlabel("Population shift from reallocation (%)")
ax.set_ylabel("Democratic voter registration (%)")
plt.show()
```
